### 1. 뉴스 제목 추출하기

```python
# IT/과학 뉴스 
#1. 직접 url을 선언할 경우
# url = 'https://news.naver.com/section/105'

req_param = {
    'sid': 105 #2. dict타입으로 요청 파라미터를 선언할 경우
}
# 
url = 'https://news.naver.com/section/{sid}'.format(**req_param) #**은 딕셔너리에 파라미터 전달하기 위해 사용
print(url)

# 요청 헤더 설정 : 브라우저 정보 (설정하는 이유는 사람처럼 보이게 하기 위함)
req_header = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)
print(res.status_code)
print(res.ok)
print(type(res))  
print(res.text) #응답객체에 들어있는 text 속성 (ex.해당 페이지 소스 보기에서 나오는 text들)
# 응답(response)이 OK 이면
# 응답 (response)에서 text 추출
# BeautifulSoup 객체 생성  
if res.ok:
    #pass        #if문에 해당하는 하위코드가 없으니 실행 시 pass 하라
    soup = BeautifulSoup(res.text,'html.parser') #parser는 정해진 틀이 있어서 html.paersr로 입력해야함
print(len(soup.select('a'))) #a라는 css 선택자를 print 해달라(len은 길이를 나타내는 함수 len 없으면 많은 a선태자 데이터가 나오기 때문에 len으로 감싸줌)
#print(len(soup.select('div.sa_text a'))) #div 밑에 있는 css 선택자 a를 print(class속성을 가진 선택자).div 속성을 함께 적어줘야함
print(len(soup.select("div.sa_text a[href*='https://n.news.naver.com/mnews/']")))
# CSS 선택자를 사용해서 a tag 목록 가져오기
a_tags = soup.select("div.sa_text a[href*='https://n.news.naver.com/mnews/']")
print(type(a_tags),type(a_tags[0])) #a_tags 타입 과 a_tags 타입 인덱스 확인 (list 형태로 가져옴-<class 'bs4.element.ResultSet'> <class 'bs4.element.Tag'>)
# <a> 태그 리스트 순회하기
for a_tag in a_tags:
    title = a_tag.text.strip()    #확인시 왼쪽으로 쓰거나 해당 print처럼 사용 가능print(a_tag.text.strip()) - strip : 공백제거
    link = a_tag['href']
    print(title,link)  
    #print(type(a_tag),a_tag) #a_tag타입과 a_tag 자체를 print로 확인
# print(soup.select("div.sa_text a[href*='mnews/article']"))
else:
# 응답(response)이 Error 이면 status code 출력    
    print(f'Error code = {res.status_code}')
```

### 2. Image 다운로드
* referer 요청 헤더를 반드시 설정해야 한다.

```python
import requests
import os #운영체제와 관련된 모듈

#육아일기 시즌2 73회차
req_header = {
    'referer':'https://comic.naver.com/webtoon/detail?titleId=812354&no=208&week=thu', #해당 만화 장면 켰을때의 url
     'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.3'
}

img_urls = [
    #개발자도구를 이용해서 이미지 주소(jpg 경로) 확인 후 입력
    'https://image-comic.pstatic.net/webtoon/812354/208/20251223155029_0d14a9e6cdd6b23342f96d1bd33d01c3_IMAG01_1.jpg',
    'https://image-comic.pstatic.net/webtoon/812354/208/20251223155029_0d14a9e6cdd6b23342f96d1bd33d01c3_IMAG01_2.jpg',
    'https://image-comic.pstatic.net/webtoon/812354/208/20251223155029_0d14a9e6cdd6b23342f96d1bd33d01c3_IMAG01_3.jpg'
]

for img_url in img_urls:
    # requests 의 get(url, headers) 함수 호출하기
    res = requests.get(img_url, headers=req_header) #headers를 설정하지 않으면 404(권한없음)에러 뜸
    print(res.status_code)    
    # binary 응답 데이터 가져오기
    img_data = res.content
    print(img_data)    
    # url에서 파일명만 추출하기
    file_name = os.path.basename(img_url)
    print(file_name)    
    # binday data를 file에 write하기(저장하기)
    with open(file_name,'wb') as file:
        print(f'Writing to {file_name}({len(img_data):,} bytes)') #파일명찍고 파일의 사이즈
        file.write(img_data)
```

## 리펙토링 된 코드

```python
import requests
from bs4 import BeautifulSoup
import os

# 기본 설정
url = 'https://comic.naver.com/webtoon/detail?titleId=833255&no=3&week=tue'
req_header = {'referer': url}
imgdir_name = 'img'

# 이미지 저장 폴더가 없으면 생성(makedirs = 하위 서브디렉토리 까지 생성)
os.makedirs(imgdir_name, exist_ok=True)

# 웹 페이지 요청 및 확인
res = requests.get(url)
if not res.ok:
    print(f'Error Code = {res.status_code}')
    exit()

# 이미지 URL 추출
soup = BeautifulSoup(res.text, 'html.parser')
img_url_list = [img_tag['src'] for img_tag in soup.select("img[src*='IMAG01']")]

# 이미지 다운로드
for img_url in img_url_list:
    res = requests.get(img_url, headers=req_header)
    if res.ok:
        img_data = res.content
        file_path = os.path.join(imgdir_name, os.path.basename(img_url))
        with open(file_path, 'wb') as file:
            print(f'Writing to {file_path} ({len(img_data):,} bytes)')
            file.write(img_data)
    else:
        print(f'Error Code = {res.status_code} for {img_url}')
```

### 4. 캡챠(이미지) API 호출하기

* 1. 캡차 키 발급 요청
  2. 캡차 이미지 요청
  3. 사용자 입력값 검증 요청

```python
#.env파일 읽어오기(해당 파일 실행 중에 .env를 만들거나 수정 하면 restart 하면 됨)
from dotenv import load_dotenv
import os

load_dotenv() #env파일을 메모리상에 적재 시켜주는 함수

#CLIENT_SECRET,CLIENT_ID 가져와서 변수에 담음
CLIENT_ID = os.getenv("CLIENT_ID")
print(CLIENT_ID[:4])

CLIENT_SECRET = os.getenv("CLIENT_SECRET")
print(CLIENT_SECRET[:4])
```

```python
# 캡차 키 발급 요청 (실행시 결과 키 값:scwWeA9fEZRO0S5v)
import urllib.request

code = "0"
url = "https://openapi.naver.com/v1/captcha/nkey?code=" + code
request = urllib.request.Request(url)
#요청 헤더 설정 (CLIENT_ID,CLIENT_SECRET = env 파일에 작성한 것 읽어오기) 
request.add_header("X-Naver-Client-Id",CLIENT_ID) 
request.add_header("X-Naver-Client-Secret",CLIENT_SECRET)

response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
# 캡차 이미지 요청
import urllib.request

key = "brXg4IKKPxGOUW3h" # 캡차 Key 값
url = "https://openapi.naver.com/v1/captcha/ncaptcha.bin?key=" + key
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",CLIENT_ID)
request.add_header("X-Naver-Client-Secret",CLIENT_SECRET)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    print("캡차 이미지 저장")
    response_body = response.read()
    with open('img/captcha.jpg', 'wb') as f: #img폴더 아래에 captcha이미지 생성 및 저장 
        f.write(response_body)
else:
    print("Error Code:" + rescode)
```

```python
#requests를 사용하는 코드로 변경
#  사용자 입력값 검증 요청
import requests

code = "1"
key = "brXg4IKKPxGOUW3h"
value = "E7P6LS9"
req_header={ 
    "X-Naver-Client-Id": CLIENT_ID,
    "X-Naver-Client-Secret" : CLIENT_SECRET
}#req_header라는 딕셔너리 생성
req_param = {
    "code" : code,
    "key" : key,
    "value" : value #이미지에서 나온 숫자 값
} # <=url = "https://openapi.naver.com/v1/captcha/nkey?code=" + code + "&key=" + key + "&value=" + value 해당 코드를 쿼리스트링으로 변경

url = "https://openapi.naver.com/v1/captcha/nkey"

#오류 날수 있을 경우를 위해 try구문 작성
try:
    res = requests.get(url, headers=req_header, params=req_param)
    res.raise_for_status() #오류(4xx,5xx)가 나면 예외를 발생시킴(발생 시 except로 이동해서 오류메시지 출력)

    print(res.text)
except requests.exceptions.RequestException as e:
    print(f'Status code : {res.status_code}')
    print(f'Error 발생 : {e}')
```

### 5. 블로그 검색하기

```python
import requests
from pprint import pprint #(from pprint(모듈이름-pprint.py) import pprint(pprint라는 모듈 안에 있는 pprint라는 함수이름) )

headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET,
}

payload = {
    'query': '파이썬', #검색어
    'display': 100, #한번에 표시할 검색 결과 개수
    'sort': 'sim' #검색 결과 정렬 방법(정확도 순으로 내림차순 정렬, 지정 안해도 디폴트 값은 sim)
}

url = 'https://openapi.naver.com/v1/search/blog.json'


# requests get(url, params, headers) 요청 
res = requests.get(url, params=payload, headers=headers)
# json() 함수로 응답 결과 가져오기 
print(len(res.json()['items']))
pprint(res.json()['items'])

items_data = res.json()['items']
# 'title' , 'bloggername' , 'description' , 'bloggerlink' , 'link'
```
