2-2. 문자열을 입력받아 대문자로 변환, 소문자로 변환, 길이 출력하는 프로그램을 작성하세요.


```python
character = input("문자열을 입력하세요 : ")

capital = character.upper() #.upper => 소문자를 대문자로 변환 하는 함수
small_letter = character.lower() # .lower => 대문자를 소문자로 변환하는 함수
length =len(character) # .len => 문자열을 길이로 변환하는 함수

print(f"대문자:{capital}")
print(f"소문자:{small_letter}")
print(f"문자열 길이:{length}")
```

2-3. 이메일 주소를 입력받아 @ 기호를 기준으로 사용자명과 도메인을 분리하여 출력하는 프로그램을 작성하세요.

```python
email = input("이메일을 입력하세요 : ")

user_name ,connector ,domain = email.partition("@") #@를 기준으로 분리(partition = 분리함수)
# @ 는 connector 변수에 저장

print(f"사용자명:{user_name}")
print(f"도메인:{domain}")
```

2-4. 문장을 입력받아 공백을 제거하고, 단어의 개수를 세는 프로그램을 작성하세요.

```python
text = input("문장을 입력하세요:")

gap = text.strip() #.strip() 함수는 중간공백은 유지 하고 앞 뒤 공백만 제거
word = text.split() #.split() 사용으로 공백을 기준으로 단어를 나누어 리스트로 생성
word_count = len(word)

print(f"공백제거:{gap}")
print(f"단어개수:{word_count}개")
```

문제 2-5. 문자열을 입력받아 특정 문자가 몇 번 나타나는지 세는 프로그램을 작성하세요.

```python
character = input("문자열을 입력하세요 : ")
find_chr = input("찾을 문자를 입력하세요:")

count = character.count(find_chr) #.count() 함수로 개수 확인

print(f"문자'{find_chr}'이 : {count}번 나타납니다.") #'{find_chr}' 로 출력시 ''도 같이 출력되도록
```
