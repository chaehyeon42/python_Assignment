문제 3-1. 5개의 숫자를 리스트에 저장하고 합계와 평균을 구하는 프로그램을 작성하세요.

```python
number = [10,20,30,40,50]

number_sum = sum(number) #sum함수로 합계구하기
number_avg = number_sum/ len(number) #총합/리스트에 저장한 숫자 개수

print(f'숫자들:{number}')
print(f'합계:{number_sum}')
print(f'평균:{number_avg}')
```

3-3. 학생들의 성적을 딕셔너리로 저장하고 평균 점수를 계산하는 프로그램을 작성하세요.

```python
exam_result = {"김철수": 85, "이영희": 92, "박민수": 78, "최지우": 95} #학생들의 성적 딕셔너리로 저장

total_sum = sum(exam_result.values()) # .values()함수로 모든점수만 추출하여 합계 계산

avg = total_sum / len(exam_result) # 점수 합계 / 학생 성적 수

print("학생 성적표:")
for name, result in exam_result.items(): #student_scores.items() -> 딕셔너리의 (키,값) 쌍을 꺼내주는 함수이용 =  name(key)과 result(value) 
    print(f"{name}: {result}점") # f-string을 사용하여 학생의 이름과 점수를 형식에 맞춰 출력

print(f"전체 평균 점수: {avg:.1f}점") #평균값은 소수점 아래 둘째 자리까지만 표시
```

3-4. 좌표를 튜플로 저장하고 두 점 사이의 거리를 계산하는 프로그램을 작성하세요.

```python
import math  # 제곱근(sqrt) 계산을 위해 수학 모듈가져오기

#좌표를 튜플로 저장
point1 =(0, 0)
point2 =(3, 4)

#두 점 사이의 거리 계산
distance = math.dist(point1,point2) #math.dist() 함수를 사용하여 두 점 사이의 거리를 한 줄로 계산

print(f"점 1:{point1}")
print(f"점 2:{point2}")
print(f"두 점 사이의 거리:{distance}")
```

3-5.쇼핑 카트를 딕셔너리로 구현하여 상품을 추가하고 총 가격을 계산하는 프로그램을 작성하세요.

```python
#중첩 딕셔너리 생성
shopping_cart = {"사과": {"count": 2, "price": 1000}, "바나나": {"count": 3, "price": 800}, "오렌지": {"count": 1, "price": 1500}}

print("쇼핑카트:")

#총액 저장 변수 초기화
total_sum = 0

#딕셔너리 반복하여 각 상품의 가격 계산 및 출력
#딕셔너리의 .items() 메서드를 사용
for item, info in shopping_cart.items():
    #수량과 단가 가져와서 해당 상품 계산
    subtotal = info["count"] * info["price"]

    #합계에 개별 물품 계산 값 더하기
    total_sum += subtotal

    #f-string을 사용하여 변수들을 문자열 사이에 작성
    print(f"{item}: {info['count']}개 (개당 {info['price']}원) = {subtotal}원")

print(f"총 가격: {total_sum}원")
```
