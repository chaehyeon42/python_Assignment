### BeautifulSoup 
* select() 함수 사용
* melon 100 chart 데이터 파싱

```python
import re #정규표현식 사용 시 작성
import requests
from bs4 import BeautifulSoup

url = 'https://www.melon.com/chart/index.htm'

headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

# 노래 상세정보 song_url = f'https://www.melon.com/song/detail.htm?songId={song_id}'
#[{},{}]
res = requests.get(url, headers=headers)
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')    
    atag_list = soup.select("a[href*='playSong']") #a태그 안에 playSong이 들어있는 경로 다 조회 (<a href="javascript:melon.play.playSong('1000002721',33241003);)
    print(len(atag_list)) 

    #100곡의 노래 목록(song_list) 저장
    song_list = []
    for idx,atag in enumerate(atag_list,1):
        #1곡 노래 목록(song_dict) 
        song_dict = {}
        #노래 제목 가져오기
        song_dict['title'] = atag.text

        #href 링크 추출
        href = atag['href'] #href = 문자열
        #print(href)        (정규식     , 정규식을 적용할 값)
        matched = re.search(r'(\d+)\);',href)#a태그 안에 문자열 안에 있는 songID 추출
        if matched:
            song_id = matched.group(1)
            # print(type(matched))
            #print(matched.group(0),matched.group(1))
        
        song_dict['id'] = song_id   
        #곡 상세 경로 딕셔너리에 저장
        song_url = f'https://www.melon.com/song/detail.htm?songId={song_id}'
        song_dict['url'] = song_url  
        print(f'song = {idx} {song_dict}')

        #song_dict를 song_list에 append
        song_list.append(song_dict)
```

### 곡상세 정보 추출하기

```python
import re
import requests
from bs4 import BeautifulSoup
from pprint import pprint

headers = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

# Song 100곡의 상세정보 목록을 저장할 list 선언
song_lyric_list = list()
print('===> 100 곡 노래 파싱 시작')
#      song = 위에서 정의한 변수
for idx,song in enumerate(song_list,1): 
    print(f'==> {idx} {song['title']}')
    # Song 1곡의 상세정보를 저장할 dict 선언
    song_lyric_dict = dict()
                                    #headers는 접근 권한때문에 반드시 정의 해줘야함.
    res = requests.get(song['url'], headers=headers)
    if res.ok:
        soup = BeautifulSoup(res.text,'html.parser')
        song_lyric_dict['곡명'] = song['title']
          #   가수제목 가져오기        ("a[href*='goArtistDetail'] span") => a태그 밑에 있는 href 안에 ArtistDetail이라는 변수가 있는 경로 찾으면서 해당 찾은 경로 아래에 span이 있는 경로를 가져오기   
        singer_span = soup.select_one("a[href*='goArtistDetail'] span")
        song_lyric_dict['가수'] = singer_span.text
        
        song_dd = soup.select('div.meta dd') #song_dd는 ResultSet타입, song_dd[0]는 Tag 타입
        if song_dd:
           #방법1. 앨범 텍스트를 가져온 후, 정규 표현식으로 특수문자와 공백 제거
           # album_raw = song_dd[0].text
           # album_clean = re.sub(r'[^a-zA-Z0-9가-힣]', '', album_raw)# [^a-zA-Z0-9가-힣] 은 모든 문자를 의미
           # song_lyric_dict['앨범'] = album_clean
           #인덱스 형식으로 작성할 때 가져온 엘리먼츠가 있는지 확인하고 사용해야함
            
            #방법2. strip함수를 이용하여 공백을 분리 시킨 후 다시 앨범제목을 붙여서 특수문자가 나오면 빈 공백 문자로 치환
            song_lyric_dict['앨범'] = song_dd[0].get_text(strip=True).replace('\xa0','')
            song_lyric_dict['발매일'] = song_dd[1].text
            song_lyric_dict['장르'] = song_dd[2].text

        #song 상세정보 url 저장
        song_lyric_dict['detail_url'] = song['url']

        # 좋아요 건수(=동적으로 렌더링 함) 가져오기 ajax_url = f'https://www.melon.com/commonlike/getSongLike.json?contsIds={song_id}'
        # -> 해당 앨범 상세 페이지에서 개발자도구로 네트워크에서 Response에 getSongLike.json 확인
        # -> json에 있는 데이터 추출하여 좋아요 건수 가져올 수 있음.
        song_id = song['id']  #앞서 딕셔너리에 저장한 id를 가져와서 지정
        ajax_url = f'https://www.melon.com/commonlike/getSongLike.json?contsIds={song_id}'
        res = requests.get(ajax_url, headers=headers)
        if res.ok:
            song_lyric_dict['좋아요'] = res.json()['contsLike'][0]['SUMMCNT'] #contsLike 리스트 안에 SUMMCNT라는 딕셔너리 하나만 추출
       
        # 노래 가사     
        lyric_div = soup.select('div#d_video_summary')
        if lyric_div:
            lyric = lyric_div[0].text #[0]을 쓴 이유는 lyric_div = soup.select 에서 select를 사용했기 때문
        else:
            lyric = ''    
        #song_lyric_dict['가사'] = lyric #딕셔너리에 lyric변수를 넣어서 출력 결과
        
        # \n\r\t 특수문자를 찾는 Pattern객체 생성 : re.compile 기능
        pattern = re.compile(r'[\n\r\t]') 
        song_lyric_dict['가사'] = pattern.sub('', lyric) # pattern = '' -> 빈 공백을 문자열로 치환
        
        #print(song_lyric_dict)

        #song list에 노래 상세정보 dict를 저장
        song_lyric_list.append(song_lyric_dict)
    else:
        print(f'Error 코드 = {res.status_code}')

print(len(song_lyric_list))
pprint(song_lyric_list[98:])
print('===>100 곡 노래 파싱 끝')
```

## MySQL 데이터베이스에서 최근 1년 이내에 발매된 곡들을 가져와서 좋아요(likes) 순으로 정렬

```python
import pandas as pd           # 데이터를 표 형태(DataFrame)로 가공하기 위한 라이브러리
import pymysql                # 파이썬에서 MySQL 데이터베이스에 접속하기 위한 드라이버
from sqlalchemy import create_engine  # 데이터베이스 연결 엔진을 관리하는 도구
from datetime import datetime, timedelta  # 날짜 계산(현재 시간, 1년 전 등)을 위한 모듈

# 1. DB 연결 설정: PyMySQL을 MySQLdb라는 이름으로 사용하도록 설정 (SQLAlchemy와의 호환성 확보)
pymysql.install_as_MySQLdb()

def get_recent_hits_to_markdown(filename="melon_trend_analysis.md"):
    """
    최근 1년치 데이터를 DB에서 가져와 마크다운 리포트로 저장하는 함수
    """
    engine = None
    conn = None
    
    # 최근 1년 전 날짜 계산 (현재 날짜에서 365일을 뺀 후 'YYYY-MM-DD' 형식으로 변환)
    one_year_ago = (datetime.now() - timedelta(days=365)).strftime('%Y-%m-%d')
    
    try:
        # 데이터베이스 연결 엔진 생성 (접속 정보: 사용자명:비밀번호@주소:포트/DB이름)
        engine = create_engine('mysql+pymysql://python:python@localhost:3306/python_db?charset=utf8mb4')
        conn = engine.connect()  # 실제로 DB 접속 수행

        # SQL 쿼리 작성: 발매일이 1년 전(%s)보다 크거나 같은 곡들을 좋아요(likes)가 많은 순서로 정렬
        sql = """
        SELECT release_date, title, singer, album, genre, likes 
        FROM songs100 
        WHERE release_date >= %s 
        ORDER BY likes DESC;
        """
        
        # pandas의 read_sql_query를 사용해 SQL 결과를 즉시 표(DataFrame) 형태로 가져옴
        # params=(one_year_ago,)는 SQL 내의 %s 자리에 날짜 데이터를 안전하게 전달함
        df = pd.read_sql_query(sql, con=conn, params=(one_year_ago,))
        
        # 만약 가져온 데이터가 없다면 메시지를 출력하고 함수 종료
        if df.empty:
            print(f"{one_year_ago} 이후 발매된 데이터가 없습니다.")
            return None

        # 2. 마크다운(.md) 파일 저장 시작 (UTF-8 인코딩으로 한글 깨짐 방지)
        #file 저장시 데이터폴더 아래에 저장
        file_path = 'data/' + filename
        with open(file_path, "w", encoding="utf-8") as f:
            # 리포트의 대제목과 정보 작성
            f.write(f"# 🎵 최근 1년 발매 곡 분석 리포트\n\n")
            f.write(f"> **조회 기준일:** {one_year_ago} 이후 ~ 현재\n\n")
            f.write(f"총 **{len(df)}**개의 곡이 검색되었습니다.\n\n")
            
            # --- [추가 기능] 데이터 요약 섹션 작성 ---
            f.write("### 📊 장르별 곡 수 요약\n")
            
            # 'genre' 컬럼의 데이터별 개수를 세고(value_counts), 이를 다시 표 형태(to_frame)로 변환
            # reset_index()를 사용해 장르 이름이 인덱스가 아닌 일반 컬럼으로 오게 함
            genre_counts = df['genre'].value_counts().to_frame().reset_index()
            
            # 결과 표의 컬럼명을 사용자가 보기 좋게 '장르'와 '곡 수'로 변경
            genre_counts.columns = ['장르', '곡 수']
            
            # 요약된 표를 마크다운 형식의 문자열로 변환하여 파일에 기록
            f.write(genre_counts.to_markdown(index=False) + "\n\n")
            
            f.write("--- \n\n") # 리포트 가독성을 위한 구분선 추가
            
            # --- 상세 목록 섹션 작성 ---
            f.write("### 📝 상세 곡 목록 (좋아요순)\n")
            # 전체 검색 결과(df)를 마크다운 표 형식으로 파일에 기록
            f.write(df.to_markdown(index=False))
            
        # 성공 메시지 출력
        print(f"✅ 분석 결과가 '{filename}'으로 저장되었습니다.")
        return df  # 필요 시 추가 활용을 위해 결과 데이터프레임 반환

    except Exception as e:
        # DB 연결이나 파일 저장 중 오류 발생 시 에러 내용 출력
        print(f"❌ 오류 발생: {e}")
    finally:
        # 에러 발생 여부와 상관없이 마지막에는 항상 DB 연결을 종료하여 자원 낭비 방지
        if conn is not None: 
            conn.close()
        if engine is not None:
            engine.dispose()

# 스크립트 실행 시작점
if __name__ == "__main__":
    # 작성한 함수를 호출하여 리포트 파일 생성
    result_df = get_recent_hits_to_markdown("melon_trend_analysis.md")
    
    # 데이터가 정상적으로 반환되었다면 완료 메시지 출력
    if result_df is not None:
        print("작업 완료! 생성된 파일을 확인해 보세요.")
```
