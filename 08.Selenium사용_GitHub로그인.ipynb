### WebDriver 의 Chrome 객체생성
* python.org 웹사이트로 요청을 보내기
* pypi 메뉴를 클릭해서 pypi.org 웹사이트로 이동하기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager #ChromeDriverManager:Chrome 용 webdriver를 띄워줌.
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.python.org/")   #해당 사이트 방문

#webdriver에 포함된 Chrome 객체를 생성 - chromedriver.exe 사용해서

#li 엘리먼트 선택해서 가져오기
#               find_elements = elements찾을때 사용 By.CSS_SELECTOR -> css 선택자 형식으로 elements를 찾는다
li_tags = driver.find_elements(By.CSS_SELECTOR,'div#top ul.menu li') 
print(li_tags)
#pypi 엘리먼트를 찾아서 저장할 변수 선언
pypi = None
for li_tag in li_tags:
    print(li_tag.text, type(li_tag))
    if li_tag.text == 'PyPI':   #해당 li태그에 text가 PyPI인지 확인
        pypi = li_tag
        
time.sleep(3)

#pypi 엘리멘트의 click() 함수 호출
pypi.click()

#5초로 waiting time 설정
time.sleep(5)
#driver 종료
driver.quit()
```

### 네이버 포털에 파이썬 검색어를 입력
* 검색어를 입력하고 엔터를 치고
* 검색된 결과중에서 지식인에 검색된 링크를 클릭

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
#네이버 포털에 요청 보내기
driver.get('https://www.naver.com/')
time.sleep(2)

#검색어 엘리먼트 가져오기
search_tag = driver.find_element(By.CSS_SELECTOR,'#query') ##query => 검색어 id값 / id값을 나타낼땐 # 입력 후 id값 작성
#print(search_tag)

#검색창에 파이썬 키워드를 입력한다.
search_tag.send_keys('python') #send_keys ->검색창에 키워드 입력한 것을 나타냄
#엔터를 입력한다.
search_tag.send_keys(Keys.ENTER)
time.sleep(5)

#검색된 결과 중에서 a tag 가져오기 <a class='lnk_tit'>
a_tags = driver.find_elements(By.CSS_SELECTOR,'ul.lst_type li a.lnk_head') #ul태그 안에 li태그 안에 a태그의 class이름이 lnk_head인 사이트 가져오기
if a_tags:
    #첫번째 a tag 선택하고, click() 함수 호출
    a_tags[0].click()

#지연 시간을 5초로 설정
time.sleep(5)
#driver 종료
driver.quit()
```

### 로그인 실패할 경우 처리한 코드

```python
#.env파일 읽어오기(해당 파일 실행 중에 .env를 만들거나 수정 하면 restart 하면 됨)
from dotenv import load_dotenv
import os

load_dotenv() #env파일을 메모리상에 적재 시켜주는 함수

#GIT_USERNAME,GIT_PASSWORD 가져와서 변수에 담음
GIT_USERNAME = os.getenv("GIT_USERNAME")
print(GIT_USERNAME[:4])

GIT_PASSWORD = os.getenv("GIT_PASSWORD")
print(GIT_PASSWORD[:4])
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

# 1. 대상 URL 설정 및 드라이버 초기화
url = 'https://github.com/login'

# 크롬 드라이버 자동 설치 및 서비스 객체 생성
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# GitHub 로그인 페이지로 이동
driver.get(url)

# 웹 리소스가 로드될 때까지 최대 2초간 대기 (모든 요소가 나타날 때까지 기다림)
driver.implicitly_wait(2)

# 2. 로그인 정보 입력
# ID 입력창을 찾아 사용자명(GIT_USERNAME) 전송
driver.find_element(By.CSS_SELECTOR,'input#login_field').send_keys(GIT_USERNAME)
# 비밀번호 입력창을 찾아 비밀번호(GIT_PASSWORD) 전송
driver.find_element(By.CSS_SELECTOR,'input#password').send_keys(GIT_PASSWORD)
# 사람처럼 보이게 하기 위해 2초간 일시 정지
time.sleep(2)

# 비밀번호 입력창에서 엔터 키를 눌러 로그인 제출
driver.find_element(By.CSS_SELECTOR,'input#password').send_keys(Keys.RETURN)

# 3. 로그인 성공 여부 판별
login_ok = None
try:         
    # 로그인 실패 시 나타나는 경고창(Flash Alert) 요소가 있는지 확인
    div_error = driver.find_element(By.CSS_SELECTOR,'div.js-flash-alert')
    print(div_error)
    if div_error:
        login_ok = False # 에러 태그가 있으면 로그인 실패
        print("login fail ----------------",login_ok)    
except NoSuchElementException as nexp:
    # 에러 태그를 찾을 수 없다면 로그인 성공으로 간주
    login_ok = True
except Exception as exp:
    # 기타 예외 발생 시 에러 메시지 출력
    print(exp)

print('login_ok', login_ok)

# 4. 로그인 성공 후 자동화 로직
if login_ok:    
    try:
        # 우측 상단 아바타(프로필 이미지) 버튼을 찾아 리스트로 가져옴
        avatar_img = driver.find_elements(By.CSS_SELECTOR, 'span.Button-label img')
        if avatar_img:
            avatar_img[0].click() # 첫 번째 아바타 이미지 클릭 (메뉴 열기)
            time.sleep(4) # 메뉴 애니메이션을 고려하여 4초 대기
            
            # 메뉴 중 'Repositories' 탭으로 가는 링크를 찾아 클릭
            repo_a_tag = driver.find_elements(By.CSS_SELECTOR,"a[href*='?tab=repositories']")
            if repo_a_tag:
                repo_a_tag[0].click()
            time.sleep(4)

            # 저장소 목록에서 첫 번째 저장소의 제목 링크를 찾아 클릭
            repo1_a_tags = driver.find_elements(By.CSS_SELECTOR,"ul li.col-12 h3.wb-break-all a")
            if repo1_a_tags:
                repo1_a_tags[0].click()
            time.sleep(4)
            
        # 5. 로그아웃 과정 시작
        # 다시 아바타 버튼을 클릭하여 메뉴를 엶
        avatar_img = driver.find_elements(By.CSS_SELECTOR, 'span.Button-label img')
        if avatar_img:
            avatar_img[0].click()
            time.sleep(4)
            
            # 'Sign out' 메뉴 링크를 찾아 클릭
            logout_a_tag = driver.find_elements(By.CSS_SELECTOR,"a[href='/logout']")
            if logout_a_tag:
                logout_a_tag[0].click()
                
        time.sleep(2)
        
        # 실제 로그아웃 버튼(최종 확인 버튼)을 찾음
        # 주의: driver.find_elements_by_name 방식은 구버전 방식이므로 에러가 날 수 있음
        logout_input = driver.find_elements(By.NAME, "commit") 
        if logout_input:
            logout_input[0].submit() # 폼 데이터를 제출하여 로그아웃 완료
        time.sleep(2)
        
    except Exception as exp:
        print(exp)

# 모든 작업 종료 후 브라우저를 닫기 전 3초 대기
time.sleep(3)
# 브라우저 종료 및 프로세스 정리
driver.quit()
```
