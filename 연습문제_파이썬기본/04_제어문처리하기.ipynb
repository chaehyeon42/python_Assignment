4-1. 점수를 입력받아 학점을 출력하는 프로그램을 작성하세요. (90이상:A, 80이상:B, 70이상:C, 60이상:D, 미만:F)

```python
#점수 입력 받는 부분, input은 문자열로 저장되므로 int()로 숫자로 바꿔줌
score = int(input("점수를 입력하세요: "))

#조건문으로 점수대별 학점 결정
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f"점수 {score}점의 학점은 {grade}입니다.")
```

4-1.1부터 100까지의 숫자 중에서 3의 배수의 합을 구하는 프로그램을 작성

```python
#3의 배수 리스트 생성
multiples_of_3 = []

#1부터 100까지 3의 배수 찾음
#range(1, 101)은 1부터 100까지의 숫자를 생성

for i in range(1, 101):
    # % 연산자는 나머지 구함. 3으로 나눈 나머지가 0이면 3의 배수
    if i % 3 == 0:
        # 배수를 찾으면 리스트에 추가
        multiples_of_3.append(i)

#리스트에 담긴 숫자 합계
total_sum = sum(multiples_of_3)

#3의 배수 개수
count = len(multiples_of_3)

print(f"1부터 100까지 3의 배수: {multiples_of_3}")
print(f"3의 배수의 합: {total_sum}")
print(f"3의 배수의 개수: {count}개")
```

4-3.구구단을 출력하는 프로그램을 작성하세요. 사용자가 입력한 단만 출력하세요.


```python
#출력하고 싶은 단을 입력
#숫자로 계산해야 되서 int()를 사용해 정수로 변환
dan = int(input("몇 단을 출력할까요? "))

print(f"\n{dan}단 구구단:\n")

#for 반복문을 사용하여 1부터 9까지 곱셈
#range(1, 10)은 1부터 9까지의 숫자 생성
for i in range(1, 10):
    # 결과 값을 계산합니다.
    result = dan * i
    # 출력 형식: 7 x 1 = 7
    print(f"{dan} x {i} = {result}")
```

4-4. 숫자를 계속 입력받아서 0을 입력할 때까지 합계를 구하는 프로그램을 작성하세요.


```python
#숫자를 누적해서 더해줄 변수를 0으로 초기화
total_sum = 0

#while 문을 사용하여 무한 루프 생성
# 'True'를 조건으로 넣어 break를 만날 때까지 계속 반복되도록
while True:
    # 숫자 입력
    num = int(input("숫자를 입력하세요 (0을 입력하면 종료): "))
    
    # 입력받은 숫자가 0인지 확인
    if num == 0:
        # 0이 입력되면 반복문을 즉시 빠져나감
        break
    
    # 0이 아니라면 입력받은 값을 합계 변수에 계속 더해줌.
    total_sum += num

print(f"\n입력된 숫자들의 합: {total_sum}")
```

4-5.사용자가 입력한 숫자가 소수인지 판별하는 프로그램을 작성하세요.

```python
#판별할 숫자를 입력
number = int(input("숫자를 입력하세요: "))

# 소수인지 판별하기 위한 상태 변수 생성
# 일단 소수라고 가정(True)하고 시작
is_prime = True

#소수는 1보다 커야함
if number <= 1:
    is_prime = False
else:
    # 2부터 (해당 숫자 - 1)까지 나누기
    # 만약 중간에 하나라도 나누어떨어지는 수가 있다면 소수가 아님
    for i in range(2, number):
        if number % i == 0:
            is_prime = False
            break  # 소수가 아님이 확정되었으므로 더 이상의 반복은 불필요

if is_prime:
    print(f"\n{number}은 소수입니다.")
else:
    print(f"\n{number}은 소수가 아닙니다.")
```
