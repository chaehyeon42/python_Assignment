7-1. 텍스트 파일에 여러 줄의 문자열을 저장하고 읽어오는 프로그램을 작성하세요.

```python
content = ["안녕하세요", "파이썬 파일 처리를 연습하고 있습니다", "오늘은 좋은 날씨입니다"]

# 파일 쓰기
with open("practice.txt", "w", encoding="utf-8") as f:
    print("파일에 저장할 내용:")
    for line in content:
        print(line)
        f.write(line + "\n")

# 파일 읽기
print("\n파일에서 읽어온 내용:")
with open("practice.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

7-2. CSV 형태의 학생 성적 데이터를 파일에 저장하고 읽어서 평균을 계산하는 프로그램을 작성하세요.

```python
import csv

data = [
    ['이름', '점수'],
    ['김철수', 85], ['이영희', 92], ['박민수', 78], ['최수진', 95]
]

# 1. CSV 파일 저장
with open('grades.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(data)
print("학생 성적이 grades.csv에 저장되었습니다.\n")

# 2. CSV 파일 읽기 및 분석
print("성적 분석 결과:")
total_score = 0
count = 0

with open('grades.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader) # 헤더 건너뛰기
    for row in reader:
        name, score = row[0], int(row[1])
        print(f"{name}: {score}점")
        total_score += score
        count += 1

print(f"전체 평균: {total_score / count}점")
```

7-3. 텍스트 파일의 단어 빈도를 계산하는 프로그램을 작성하세요.


```python
# 분석할 텍스트 내용 
text ="파이썬 프로그래밍 언어 배우기 파이썬 언어는 쉬운 언어입니다 파이썬 프로그래밍은 매우 강력한 도구입니다"

# 파일에 저장
with open("word_frequency.txt", "w", encoding="utf-8") as f:
    f.write(text)

# 파일 읽기 및 단어 빈도 분석
word_count = {}
with open("word_frequency.txt", "r", encoding="utf-8") as f:
    words = f.read().split()
    for word in words:
        # 단어 정제 (이미지 결과에 맞춰 '언어는', '언어입니다' 등을 개별 단어로 처리)
        word_count[word] = word_count.get(word, 0) + 1

# 내림차순 정렬
print("단어 빈도 분석 결과:")
sorted_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_words:
    # 이미지의 출력 형식인 '단어: N번'에 맞춤
    if count >= 1:
        print(f"{word}: {count}번")
```

7-4.JSON 형태의 데이터를 파일에 저장하고 읽어오는 프로그램을 작성하세요.

```python
import json

# 저장할 데이터 정의 (Python 딕셔너리 형태)
data = {
    "이름": "김철수",
    "나이": 25,
    "직업": "개발자",
    "취미": ["독서", "영화감상", "코딩"],
    "주소": "서울시 강남구"
}

# JSON 파일로 데이터 저장
file_path = "data.json"
with open(file_path, "w", encoding="utf-8") as f:
    # ensure_ascii=False를 설정해야 한글이 깨지지 않고 저장됩니다.
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"데이터가 {file_path}에 저장되었습니다.\n")


# JSON 파일로부터 데이터 읽기
with open(file_path, "r", encoding="utf-8") as f:
    loaded_data = json.load(f)


print("JSON에서 읽어온 데이터:")
print(f"이름: {loaded_data['이름']}")
print(f"나이: {loaded_data['나이']}")
print(f"직업: {loaded_data['직업']}")
print(f"취미: {loaded_data['취미']}")
print(f"주소: {loaded_data['주소']}")
```

7-5. 로그 파일을 생성하고 특정 레벨의 로그만 필터링하여 출력하는 프로그램을 작성하세요.


```python
# 로그 파일 생성
log_content = """2025-07-20 09:15:00 - WARNING - 메모리 사용량이 높습니다
2025-07-20 10:30:00 - ERROR - 데이터베이스 연결 실패
2025-07-20 11:00:00 - INFO - 서버가 정상 가동 중입니다
2025-07-20 11:45:00 - ERROR - 파일을 찾을 수 없음
2025-07-20 12:00:00 - WARNING - 디스크 공간 부족"""

file_name = "system.log"

with open(file_name, "w", encoding="utf-8") as f:
    f.write(log_content)

print("로그 파일이 생성되었습니다.\n")


# 파일 읽기 및 필터링 함수 정의
def print_logs_by_level(filename, level_name):
    print(f"{level_name} 레벨 로그들:")
    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            # 해당 줄에 특정 레벨 문자열이 포함되어 있는지 확인
            if f"- {level_name} -" in line:
                print(line.strip())
    print()  # 가독성을 위한 줄바꿈


print_logs_by_level(file_name, "ERROR")
print_logs_by_level(file_name, "WARNING")
```
