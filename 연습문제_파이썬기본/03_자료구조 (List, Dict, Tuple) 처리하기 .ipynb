3-1. 5개의 숫자를 리스트에 저장하고 합계와 평균을 구하는 프로그램을 작성하세요.

```python
#숫자들: [10, 20, 30, 40, 50]

num = [10, 20, 30, 40, 50]

#숫자 합계
num_sum = sum(num)

#평균
num_avg = num_sum/ len(num)

print(f"숫자들:{num}")
print(f"합계:{num_sum}")
print(f"평균:{num_avg}")
```

3-2.과일 이름들이 담긴 리스트에서 특정 과일을 검색하는 프로그램을 작성

```python
# ['사과', '바나나', '오렌지', '포도', '딸기']
fruits = ['사과', '바나나', '오렌지', '포도', '딸기']

fruit_search = input("찾을 과일을 입력하세요:")

#입력한 과일이 있는지 확인
if fruit_search in fruits:
    # 리스트에 과일이 있을 경우 출력되는 메시지입니다.
    print(f"'{fruit_search}'가 목록에 있습니다!")
else:
    # 리스트에 과일이 없을 경우 출력되는 메시지입니다.
    print(f"'{fruit_search}'가 목록에 없습니다.")
```

3-3. 학생들의 성적을 딕셔너리로 저장하고 평균 점수를 계산하는 프로그램을 작성하세요.

```python
#학생들 성적
student_result = {"김철수": 85, "이영희": 92, "박민수": 78, "최수진": 95}

print("학생 성적 : ")

#딕셔너리의 모든 항목을 반복문(for)으로 출력
# .items()를 쓰면 이름과 점수를 동시에 가져옴
for name, score in student_result.items():
    print(f"{name}: {score}점")

#평균 점수를 계산
average = sum(student_result.values()) / len(student_result)

print(f"\n평균 점수: {average}점")
```

3-4. 좌표를 튜플로 저장하고 두 점 사이의 거리를 계산하는 프로그램을 작성

```python
# 제곱근(sqrt) 계산을 위해 수학 모듈 가져옴
import math

#튜플 형태() 로 저장
point1 = (0, 0)
point2 = (3, 4)

#두 점 사이 거리 계산
# math.sqrt()는 제곱근을, **2는 제곱을 의미
distance = math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)


print(f"점1: {point1}")
print(f"점2: {point2}")
print(f"두 점 사이의 거리: {distance}")
```

3-5.쇼핑 카트를 딕셔너리로 구현하여 상품을 추가하고 총 가격을 계산하는 프로그램을 작성하세요

```python
#쇼핑 카트 딕셔너리로 구현
shopping_cart = {"사과": [2, 1000], "바나나": [3, 800], "오렌지": [1, 1500]}

print("쇼핑 카드:")

# 총 가격을 누적해서 더해줄 변수를 0으로 초기화
total_price = 0

#딕셔너리의 아이템들을 하나씩 꺼내어 계산하고 출력
# item에는 상품명이, info에는 [수량, 단가] 리스트
for item, info in shopping_cart.items():
    count = info[0]  # 수량
    price = info[1]  # 단가
    
    # 해당 상품의 소계(수량 * 단가)를 계산
    subtotal = count * price
    
    # 전체 총 가격에 소계를 더함
    total_price += subtotal
    
    # 출력 형식: 상품명: n개 (개당 n원) = n원
    print(f"{item}: {count}개 (개당 {price}원) = {subtotal}원")

#모든 상품의 계산이 끝나면 최종 합계를 출력합
print(f"총 가격: {total_price}원")
```

3-6.리스트에서 최댓값, 최솟값, 두 번째로 큰 값을 찾는 프로그램을 작성하세요.

```python
#리스트에 저장
numbers = [15, 3, 27, 8, 19, 12, 31]

#최댓값과 최솟값
max_val = max(numbers)
min_val = min(numbers)

#리스트 내림차순 정렬
# sorted() 함수에 reverse=True 옵션을 주면 [큰 값 -> 작은 값] 순서로 정렬
sorted_numbers = sorted(numbers, reverse=True)

# 정렬된 리스트에서 인덱스 1번(두 번째 위치)을 선택
# 파이썬은 숫자를 0부터 세기 때문에 1번이 두 번째 값이 됨
second_max = sorted_numbers[1]

print(f"숫자 목록: {numbers}")
print(f"최댓값: {max_val}")
print(f"최솟값: {min_val}")
print(f"두 번째로 큰 값: {second_max}")
```

3-7.단어들이 담긴 리스트에서 가장 긴 단어와 가장 짧은 단어를 찾는 프로그램을 작성하세요.


```python
#리스트 작성
words = ['cat', 'elephant', 'dog', 'butterfly', 'ant']

# key=len 옵션을 사용하면 단어의 '길이'를 기준으로 최댓값과 최솟값 찾음
longest_word = max(words, key=len)
shortest_word = min(words, key=len)

print(f"단어 목록: {words}")

# len() 함수를 사용하여 단어 옆에 글자 수도 표시
print(f"가장 긴 단어: {longest_word} ({len(longest_word)}글자)")
print(f"가장 짧은 단어: {shortest_word} ({len(shortest_word)}글자)")
```

3-8.중복된 요소가 있는 리스트에서 중복을 제거하고 정렬하는 프로그램을 작성하세요.


```python
#원본 리스트 생성
original_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]

#set()을 사용하여 리스트의 중복 제거, set 함수는 중복된 값을 허용하지 않음
unique_set = set(original_list)

#중복이 제거된 집합을 다시 리스트로 변환하고 정렬, sorted 함수는 데이터를 오름차순(작은 순서)으로 정렬하여 리스트로 반환
sorted_list = sorted(unique_set)

print(f"원본 리스트: {original_list}")
print(f"중복 제거 후: {sorted_list}")
```

3-9.두 개의 리스트를 병합하고 공통 요소를 찾는 프로그램을 작성하세요.



```python
#두개 리스트 생성
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

#두 리스트를 병합하는데 중복을 제거하여 정렬
# set을 사용하여 두 리스트를 더한 뒤 중복을 없애고, sorted로 다시 정렬
union_list = sorted(list(set(list1 + list2)))

# set() 자료형의 '&' 연산자를 사용하여 교집합 구함
common_elements = sorted(list(set(list1) & set(list2)))

print(f"리스트1: {list1}")
print(f"리스트2: {list2}")
print(f"병합된 리스트: {union_list}")
print(f"공통 요소: {common_elements}")
```

3-10. 학생 정보를 딕셔너리의 리스트로 저장하고 나이 순으로 정렬하는 프로그램을 작성하세요.


```python
#학생 정보 딕셔너리 리스트로 저장
students = [{"이름": "김철수", "나이": 20, "전공": "컴퓨터공학과"}, {"이름": "이영희", "나이": 22, "전공": "영어영문학과"}, {"이름": "박민수", "나이": 21, "전공": "경영학과"}, {"이름": "최수진", "나이": 23, "전공": "수학과"}]

#나이(age)를 기준으로 리스트를 정렬
# key=lambda x: x["나이"]는 각 학생 딕셔너리에서 '나이' 값을 꺼내서 비교하라는 의미
sorted_students = sorted(students, key=lambda x: x["나이"])

print("나이 순으로 정렬된 학생 목록:\n")

#반복문을 통해 정렬된 리스트의 각 학생 정보를 하나씩 꺼내 출력
for s in sorted_students:
    # f-string을 사용하여 이름, 나이, 전공을 형식에 맞춰 표시
    print(f"{s['이름']} ({s['나이']}세) - {s['전공']}")
```
