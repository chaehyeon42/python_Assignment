6-1. 두 수를 받아서 사칙연산을 수행하는 함수들을 작성하는 프로그램을 만드세요.

```python
# 사칙연산 함수 정의
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    # 단순 나눗셈을 수행
    return a / b

# 사용자로부터 두 수 입력받기
num1 = int(input("첫 번째 숫자를 입력하세요: "))
num2 = int(input("두 번째 숫자를 입력하세요: "))

# 결과 출력
print(f"{num1} + {num2} = {add(num1, num2)}")
print(f"{num1} - {num2} = {subtract(num1, num2)}")
print(f"{num1} * {num2} = {multiply(num1, num2)}")
print(f"{num1} / {num2} = {divide(num1, num2)}")
```

6-2.팩토리얼을 계산하는 재귀함수와 반복문을 사용한 함수를 각각 작성하세요.

```python
# 재귀 함수를 이용한 방식
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)

# 반복문을 이용한 방식
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


test_cases = [5, 7]

for n in test_cases:
    print(f"{n}! (재귀) = {factorial_recursive(n)}")
    print(f"{n}! (반복) = {factorial_iterative(n)}")
```

6-3. 기본값 매개변수를 사용하여 인사말을 생성하는 함수를 작성하세요.

```python
#기본값 매개변수: 함수 정의 시 message="안녕하세요"와 같이 값을 미리 할당하면, 함수 호출 시 해당 인자를 생략
def greet(name, message="안녕하세요", suffix="님!"):
    return f"{message}, {name}{suffix}"

# 기본 메시지 사용
print(greet("김철수"))

# 다른 메시지 전달 (위치 인자)
print(greet("John", "Hello", "!"))

# 특정 매개변수 지정 전달 (키워드 인자)
print(greet("이영희", suffix="님! 좋은 하루 되세요!"))
```

6-5. 람다 함수를 사용하여 리스트를 다양한 방식으로 정렬하는 프로그램을 작성하세요.

```python
data = [('김철수', 85), ('이영희', 92), ('박민수', 78), ('최수진', 95)]

# 이름순, 점수순, 점수 내림차순 정렬
print(f"이름순: {sorted(data, key=lambda x: x[0])}")
print(f"점수순: {sorted(data, key=lambda x: x[1])}")
print(f"점수 내림차순: {sorted(data, key=lambda x: x[1], reverse=True)}")
```
