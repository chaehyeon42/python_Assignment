문제 1-1. 사용자로부터 두 개의 정수를 입력받아 사칙연산(+, -, *, /)의 결과를 출력하는 프로그램을 작성하세요.

```python
#  두 개의 정수 입력 받기
num1 = int(input("첫 번째 정수 입력:"))
num2 = int(input("두 번째 정수 입력:"))

# 사칙연산 수행 및 결과 출력
print(f"{num1} + {num2} = {num1 + num2}")
print(f"{num1} - {num2} = {num1 - num2}")
print(f"{num1} * {num2} = {num1 * num2}")

# 나눗셈의 경우 0으로 나누는 예외 처리 포함
if num2 != 0:
    print(f"{num1} / {num2} = {num1 / num2:.2f}")
else:
    print("0으로 나눌 수 없습니다.")
```

문제 1-2. 원의 반지름을 입력받아 원의 넓이와 둘레를 계산하는 프로그램을 작성하세요. (π = 3.14159 사용)

```python
# 파이 상수 정의
pi = 3.14159

#원의 반지름 입력 받기
r = float(input("원의 반지름 입력: "))

#원의 넓이
extent = pi * (r**2) # pi x r의 제곱
print(f"원의 넓이 : {extent:.2f}")

#원의 둘레
circum = 2 * pi *r
print(f"원의 둘레 : {circum:.2f}")
```

문제 1-3. 섭씨 온도를 입력받아 화씨 온도로 변환하는 프로그램을 작성하세요. (화씨 = 섭씨 * 9/5 + 32)

```python
Celsius = int(input("섭씨온도 입력:"))

Fahren = Celsius * 9/5 + 32

print(f"화씨 온도: {Fahren:.2f}")
```

문제 1-4. 직사각형의 가로와 세로 길이를 입력받아 넓이와 둘레를 계산하는 프로그램을 작성하세요.


```python
length = int(input("직사각형 가로 길이 입력 :"))
Height = int(input("직사각형 세로 길이 입력 :"))

weight = (length * Height)
perimeter = ((length + Height ) * 2 )

print(f"직사각형 넓이 : {weight}")
print(f"직사각형 둘레 :{perimeter}")
```
