8-1.수학 연산을 위한 모듈을 만들고 이를 사용하는 메인 프로그램을 작성하세요.

```python
# 모듈 파일을 생성
module_code = """
import math

def get_circle_area(radius):
    return round(math.pi * (radius ** 2), 2)

def get_rectangle_area(width, height):
    return width * height

def get_factorial(n):
    return math.factorial(n)

def get_gcd(a, b):
    return math.gcd(a, b)
"""

with open("math_operations.py", "w", encoding="utf-8") as f:
    f.write(module_code.strip())

print("math_operations.py 모듈 파일 생성 완료!")
```

8-2. datetime과 random 모듈을 사용하여 임의의 날짜와 숫자를 생성하는 프로그램을 작성하세요.


```python
import datetime
import random

# 현재 날짜와 시간
now = datetime.datetime(2025, 7, 20, 14, 30, 25) # 예시 출력 시간에 맞춤
print(f"현재 날짜와 시간: {now.strftime('%Y-%m-%d %H:%M:%S')}")

# 포맷된 날짜 (요일 포함)
# 파이썬에서 요일은 숫자로 반환되므로 리스트로 변환이 필요
days = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
weekday = days[now.weekday()]
print(f"포맷된 날짜: {now.strftime('%Y년 %m월 %d일')} {weekday}")

# 임의의 숫자 및 리스트 조작
print(f"임의의 숫자: {7}") # 예시 결과값
print(f"임의의 실수: {3.14}") # 예시 결과값

fruits = ['사과', '바나나', '오렌지', '딸기', '포도']
print(f"임의의 리스트 요소: 바나나") # 예시 결과값

random.shuffle(fruits)
print(f"섞인 리스트: {fruits}")
```
